```python
import pandas as pd
import numpy as np
from pathlib import Path
```

```python
# Read and load School and Student Data files into dataframes

school_Data = pd.read_csv("Resources/schools_complete.csv")
student_Data = pd.read_csv("Resources/students_complete.csv")
 
district_data_df = pd.DataFrame()  
school_Data.head()
student_Data.head()
  
  
```

```python
# district summary

# count of Unique school names
unique_schools1 = school_Data["school_name"].unique().size

# total no. of students
totalstudents = school_Data["size"].sum()

# total budget
totalbudget = school_Data["budget"].sum()

# average math and reading score
averagemathscore = student_Data["math_score"].mean()
averagereadingscore = student_Data["reading_score"].mean()


# percentage of students passing math
mathpasscount = len(student_Data[(student_Data["math_score"] >= 70)])
percentagemathpass = (mathpasscount/totalstudents) * 100



# percentage of students passing reading
readingpasscount = len(student_Data[(student_Data["reading_score"] >= 70)])
percentagereadingpass = (readingpasscount/totalstudents) * 100


# percentage of students passing maths and reading
bothpasscount = len(student_Data[(student_Data["reading_score"] > 70) & 
             (student_Data["math_score"]> 70)])
percentagebothpass = bothpasscount/totalstudents * 100




schoolSummary = pd.DataFrame({'Total No. of Unique School': [unique_schools1], 
                              'Total No. of Students': [totalstudents],
                        'Total Budget': [totalbudget], 
                        'Average Math Score': [averagemathscore],
                         'Average Reading Score': [averagereadingscore],
                          '% Passing Math': [percentagemathpass], 
                          '% Passing Reading': [percentagereadingpass],
                           '% Overall Passing':[percentagebothpass]})

#schoolSummary["Total Students"] = schoolSummary["Total Students"].map("{:,}".format)
schoolSummary["Total Budget"] = schoolSummary["Total Budget"].map("${:,.2f}".format)
schoolSummary

print(schoolSummary)


```
